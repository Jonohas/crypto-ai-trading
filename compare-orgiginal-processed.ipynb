```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_original = pd.read_csv('Data/raw/VETEUR.csv')
df_processed = pd.read_csv('Data/dataset/VETEUR.csv')
```

```python
# remove first line of original data
df_original = df_original.iloc[1:]

# select first 20 values
df_original = df_original.head(20)
df_processed = df_processed.head(20)
```

```python
# create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

# set minute column based on event_time
df_original['minute'] = df_original['event_time'].str.split(':').str[1]

# create a chart for original close price
ax1.plot(df_original["event_time"], df_original['close'], color='blue', label='Original')

# create a chart for processed close price on seperate chart
ax2.plot(df_processed['event_time'], df_processed['close'], color='red', label='Processed')
ax2.axhline(y=0, c="yellow", label="y=0")
```
