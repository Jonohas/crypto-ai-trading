```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_original = pd.read_csv('Data/raw/VETEUR.csv')
df_processed = pd.read_csv('Data/dataset/VETEUR.csv')
```

```python
# remove first line of original data
df_original = df_original.iloc[3:]

# select first 20 values
df_original = df_original.head(20)
df_processed = df_processed.head(20)
```

```python
# create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

# create a chart for original close price
ax1.plot(df_original["event_time"], df_original['close'], color='blue', label='Original close value')

# create a chart for processed close price on seperate chart
ax2.plot(df_processed['event_time'], df_processed['close'], color='red', label='Processed and normalized close value')
ax2.axhline(y=0, c="green", label="y=0")

fig.legend()
```

```python
difference_org = df_original['close'].diff()
difference_proc = df_processed['close'].diff()

print(difference_org, difference_proc)
```
